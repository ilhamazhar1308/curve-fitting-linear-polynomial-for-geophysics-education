#**LINEAR REGRESSION**

```python
import numpy as np
import matplotlib.pyplot as plt

# Input data
n = int(input('Enter the number of data points to be input: '))
s = []
5
# Input depth data
x = []
for i in range(n):
  X = float(input('Enter depth data number-%d: ' %i))
  x.append(X)

# Input temperature data
y = []
for i in range(n):
  Y = float(input('Enter temperature data number-%d: ' %i))
  y.append(Y)

print('Data X : ', X)
print('Data Y : ', Y)

xsum = 0
ysum = 0
x2sum = 0
jumlah = 0

for i in range(n):
  xsum = xsum + x[i]
  ysum = ysum + y[i]
  x2sum = x2sum + (x[i]**2)
  jumlah = jumlah + x[i]*y[i]

a = (n*jumlah - xsum*ysum)/(x2sum*n - xsum*xsum)
b = (x2sum*ysum - xsum*jumlah)/(x2sum*n - xsum*xsum)
Y = np.zeros(n)
for i in range(n):
  Y[i] = a*x[i] + b

print("The Linear Fitting is : ", Y)
print("The linear fitting function is : T =", a, "Z +", b)
plt.plot(x, y, 'o', x, Y)
plt.title('Temperature Variation with Depth')
plt.xlabel('Depth (m)')
plt.ylabel('Temperature (Celsius)')
plt.show()
```

#**POLYNOMIAL REGRESSION**

```python
import numpy as np
import matplotlib.pyplot as plt

n = int(input('Enter the number of data points to be input: '))
x = []
for i in range(n):
    data_x = float(input('Enter the value of x-axis number-%d: ' %i))
    x.append(data_x)

y = []
for i in range(n):
    data_y = float(input('Enter the value of y-axis number-%d: ' %i))
    y.append(data_y)

print('Data X:', x)
print('Data Y:', y)
polinom = int(input('Enter the polynomial degree to be used: '))

z = np.polyfit(x, y, polinom)
f = np.poly1d(z)
print('The fitted equation for degree %d: \n' %polinom, f)
x_pol = np.linspace(x[0], x[-1], 50)
y_pol = f(x_pol)

plt.plot(x, y, 'o', x_pol, y_pol)
plt.title('Depth (m)')
plt.xlabel('Depth (m)')
plt.ylabel('Temperature Variation with Depth')
plt.show()
```
